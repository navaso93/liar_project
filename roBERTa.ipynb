```python
from transformers import RobertaTokenizerFast
from transformers import RobertaForSequenceClassification
from transformers import pipeline
```

```python
def roberta(df):
    '''untrained roBERTa Model to classify statements into 3 categories , Input = Dataframe Column with statement'''
    # initialise tokenizer
    tokenizer = RobertaTokenizerFast.from_pretrained("FacebookAI/roberta-base")

    # define model with 3 categories
    model = RobertaForSequenceClassification.from_pretrained(
        "FacebookAI/roberta-base",
        num_labels=3,
        id2label={
            0: "CLASS_1",
            1: "CLASS_2",
            2: "CLASS_3",
        },
        label2id={
            "CLASS_1": 0,
            "CLASS_2": 1,
            "CLASS_3": 2,
        }
    )

    # define pipeline with truncation due to model restrictions
    pipe = pipeline(
        task="text-classification",
        model=model,
        tokenizer=tokenizer,
        truncation=True,
        max_length=512
    )

    result = pipe(df["statement"].tolist())

    return result
```
